- begin_month : 신용카드 발급 월 (부호 변환해서, 신용카드 가진 개월 수)

```python
def summary(df_train):
    frame_info = []
    for col in df_train.columns:
        # column, column unique 길이, column null 갯수, column 데이터 형태, column 데이터
        frame_info.append([col, len(df_train[col].unique()), df_train[col].isnull().sum(), df_train[col].isnull().sum()/len(df_train), df_train[col].dtype, df_train[col].unique()[:5]])

# frame_info를 dataframe 형태로 변환
    df_info = pd.DataFrame(frame_info, columns=['name', 'num_of_unique', 'num_of_nan', 'ratio_of_nan', 'dtype', 'front5_values'])
    return df_info
```

```python
# index, flag_mobiil, occyp_type

train = train.drop(['index'], axis=1)
test = test.drop(['index'], axis=1)

train = train.drop(['FLAG_MOBIL'], axis=1)
test = test.drop(['FLAG_MOBIL'], axis=1)

train = train.drop(['occyp_type'], axis=1)
test = test.drop(['occyp_type'], axis=1)
```

```python
## Binary_category

# value 값 변환

gender_order = {'F':0, 'M':1}
train['gender'] = train['gender'].map(gender_order)
test['gender'] = test['gender'].map(gender_order)

yesno_order = {'Y':1, 'N':0}
train['car'] = train['car'].map(yesno_order)
test['car'] = test['car'].map(yesno_order)

train['reality'] = train['reality'].map(yesno_order)
test['reality'] = test['reality'].map(yesno_order)

# 타입 변환

# train = train.astype({'work_phone': 'object'})
# test = test.astype({'work_phone': 'object'})

# train = train.astype({'phone': 'object'})
# test = test.astype({'phone': 'object'})

# train = train.astype({'email': 'object'})
# test = test.astype({'email': 'object'})
```

- 수치형

```python
from scipy.stats import norm, skew #for some statistics
numeric_cols = train.dtypes[(train.dtypes != "object") & (train.columns != "credit")].index

# Check the skew of all numerical features
skewed_feats = train[numeric_cols].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(10)
```

```python
# 양수는 log화
# 음수는 지수화로 변경

skewness1 = skewness[skewness > 0.75]
print("There are {} skewed numerical features to log transform".format(skewness1.shape[0]))

skewed_features1 = skewness1.index
for feat in skewed_features1:
    train[feat] = np.log1p(train[feat])
```

```python
skewness2 = skewness[skewness < -0.75]
print("There are {} skewed numerical features to exp transform".format(skewness2.shape[0]))

skewed_features2 = skewness2.index
print(skewed_features2)
for feat in skewed_features2:
    train[feat] = np.exp(train[feat])
```
