```python
from haversine import haversine
```

```python
market1 = market.iloc[:,[0,1]]
container1 = container.iloc[:,[0,1]]
```

```python
distance = [[] for _ in range(len(market1))]
for i in range(len(market1)) :
    for j in range(len(container1)) :
        distance[i].append(haversine(market1.iloc[i,], container1.iloc[j]))
```

```python
distance
```

```python
result = []
for i in range(len(distance)) :
    result.append(min(distance[i]))
```

```python
result
```

```python
where = []
for i in range(len(distance)) :
    where.append(distance[i].index(result[i]))
```
