4) 크롤링

```python
driver = webdriver.Chrome("C:/chromedriver.exe")
```

### 유튜브 크롤링

- 가수 + 제목 검색 시, 해당 Title과 Url 추출

```python
def get_youtube_with_keyword(keyword) :
    
    search_keyword_encode = requests.utils.quote(keyword)
    url = "https://www.youtube.com/results?search_query=" + search_keyword_encode

    driver.get(url)

    elements = driver.find_elements_by_css_selector('h3.title-and-badge.style-scope.ytd-video-renderer')
    href = ''
#     title = ''

    for idx, element in enumerate(elements) :
        if idx == 0 :
            href = element.find_elements_by_tag_name('a')[0].get_attribute('href')
#             title = element.find_elements_by_tag_name('a')[0].get_attribute('title')

    return href
```

```python
data
```

```python
href = []

for i in tqdm_notebook(range(start,end)) :
    
    result = get_youtube_with_keyword(data['search'][i])
    href.append(result)
```

<br>

### 멜론 크롤링

- 앨범 아이디 검색 시, 해당 이미지 추출

```python
def get_melon_with_keyword(keyword) :
    
    search_keyword_encode = str(keyword)
    url = "https://www.melon.com/album/detail.htm?albumId=" + search_keyword_encode 
    
    driver.get(url)
    
    elements = driver.find_elements_by_css_selector('a.image_typeAll')
    imgURL = ''

    for idx, element in enumerate(elements) :
        if idx == 0 :
            imgURL = element.find_elements_by_tag_name('img')[0].get_attribute('src')

    return imgURL
```

```python
img = []

for i in tqdm_notebook(range(start,end)) :
    
    result = get_melon_with_keyword(data['album_id'][i])
    img.append(result)
```
