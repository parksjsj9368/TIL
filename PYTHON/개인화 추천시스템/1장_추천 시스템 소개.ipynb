# 1장. 추천 시스템의 소개

# 추천 시스템

**추천 시스템 : 사용자의 과거 행동 데이터나 다른 데이터를 바탕으로 사용자에게 필요한 정보나 제품을 골라서 제시해 주는 시스템**

- 전통 방식 : 고객들을 취향이 비슷한 집단으로 나누고 각 집단에 잘 맞는 제품이나 서비스를 골라주는 것
- 최근 방식(개인화 추천) : 기술의 발전으로 집단의 크기를 아주 작게 줄이면, 각 집단이 한 사람이 된다. 이렇게 각 사용자별 맞춤형 추천과 서비스를 제공

책에서 사용되는 코드와 데이터 : www.crbooks.co.kr

# 1.1 주요 추천 알고리즘

## 1. 협업 필터링 (Collaborative Filtering, CF)

<br>

**특징**

- 구매, 소비한 제품에 대한 각 **소비자의 평가를 받아서 평가 패턴이 비슷한 소비자를 한 집단**으로 보고 그 집단에 속한 소비자들의 취향을 활용한 기술

- 사람들의 취향이 뚜렷이 구분되는 제품(영화, 음악, 패션 등)을 추천할 때, 더욱 정확

<br>

**한계점**

1) 소비자들의 평가 정보를 구하기 어렵다

- 이에 따라, 간접적인 정보(어떤 제품을 계속적으로 정보 검색, 어떤 제품을 쇼핑카트에 넣는것 등을 관심의 표현)를 사용

- 온라인에서는 클릭스트림 형태로 수집되어, 이를 분석해 소비자의 취향이나 니즈를 파악

2) 어떤 소비자가 어떤 제품을 좋아할지에 대한 분석은 가능하나, 왜 그 제품을 좋아할 것인가?에 대해서는 답 해줄 수 없다

<br>

**대표적인 예 : 아마존의 제품 추천, 넷플릭스 영화 추천**


## 2. 내용 기반 필터링 (Content-Based filtering, CB)

<br>

**특징**

- **제품의 내용(텍스트 정보)을 분석**해서 추천하는 기술

- 어떤 소비자가 현재 관심이 있는 것으로 보이는 책이나 뉴스가 있다면, 그 책과 뉴스의 키워드를 추려내고 다른 책이나 뉴스 중에서 비슷한 키워드를 가진 것을 찾아내서 추천

<br>

**한계점**

1) 어떤 소비자가 어떤 제품을 좋아할지에 대한 분석은 가능하나, 왜 그 제품을 좋아할 것인가?에 대해서는 답 해줄 수 없다


## 3. 지식 기반 필터링 (Knowledge-Based filtering, KB)

<br>

**특징**

- 특정 분야 전문가의 도움을 받아서 그 분야에 대한 전체적인 지식구조를 만들어서 이를 활용하는 방법

- 그 분야의 중요한 개념을 가지고 체계도를 만든다. 작성된 체계도를 바탕으로 소비자가 구매한, 혹은 관심 있는 제품과 관련이 있는 제품을 이 체계도에서 찾아서 제시

- 전체적인 구조를 알 수 있고, 지식이 중요한 분야(교육, 와인, 커피 등)에서 효과적

<br>

**한계점**

1) 각 분야에 대해서 전문가가 필요

2) 전문가의 능력에는 한계가 있기 떄문에 모든 세세한 분야에 다 적용할 수 없다


## 4. 딥러닝 (Deep Learning, DL) 추천 기술

<br>

**특징**

- AI 알고리즘 중에 현재 가장 많이 사용되는 딥러닝 방법을 사용하는 방식

- 입력으로 다양한 사용자와 아이템의 특징값을 사용하고, 이를 통해 얻은 출력은 각 아이템에 대한 각 사용자의 예상 선호도로 사용

- 각 사용자에 대해서 많은 아이템의 예상 선호도를 계산 할 수 있고, 이 중 예상 선호도가 높은 아이템을 추천

- 다양한 입력 변수를 사용 가능

**한계점**

- DL 기술은 이미지 인식 등에서 탁월한 성능을 발휘하지만, 연속값(1~5점)으로 표시되는 아이템에 대한 사용자의 선호도를 활용한 DL 추천 기술은 정확도가 다른 알고리즘에 비해서 크게 우월하지는 않다고 알려져 있다

- 다른 알고리즘과 같이 사용되는 경우가 많다


## 5. 하이브리드 (Hybrid) 기술

<br>

**특징**

- 실제로 추천 시스템을 구출할 때는 한가지만 사용하는 경우는 별로 없고, 대부분 두 가지 이상의 기술을 혼합해서 사용하는 하이브리드 형태

- 하나의 기술을 사용할 때보다 복수의 기술을 사용하는 경우에 정확도가 향상


# 1.2 추천 시스템 적용 사례

## 1. 넷플릭스의 사례

<br>

**협업 필터링 추천 시스템을 성공적으로 사용한 사례**

넷플릭스의 고객은 영화나 드라마를 보고 나서 콘텐츠에 대한 평가를 사이트에서 입력 할 수 있다

이 평가를 바탕으로 각 고객에게 맞는 콘텐츠를 추천해 준다

실제로 고객이 넷플릭스에서 소비하는 영화 중 80%가 추천에 의한 것

고객이 넷플릭스에서 영화를 많이 보면 고객의 평가 데이터가 쌓여서 더 정화한 추천을 해주기 때문에, 추천 시스템은 고객이 온라인 미디어 스트리밍 서비스 회사를 다른 회사로 쉽게 바꾸지 못하도록 만드는 역할을 한다


## 2. 아마존의 사례

<br>

**협업 필터링 추천 시스템을 성공적으로 사용한 사례**

사용자의 제품에 대한 평가 데이터를 다양한 방법으로 수집한다.

<br>

[ 직접적인 평가 ]

아마존 사용자들은 자신이 구입한 제품에 대해서 직접적인 평가(별5개)를 할 수도 있고, 글로 평가 후기를 남길 수 있다.

[ 암묵적인 평가 ]

어떤 제품의 설명 웹 페이지를 방문했는지, 방문했다면 얼마나 오래 머물렀는지, 어떤 제품을 쇼핑카트에 넣었는지, 구매했는지 등 모두 사용자의 선호를 나타내는 정보

다양한 정보들은 고객의 선호도를 계산할 때 가중치로서 사용된다.

<br>

아마존의 추천 시스템은 넷플릭스의 추천 시스템보다 더 복잡하고 어려운 시스템

첫째, 아마존은 취급하는 제품의 종류가 다양하고 사용자 수도 더 많다.

둘째, 넷플릭스보다 아마존의 시스템이 처리하는 선호 정보의 종류가 더 다양하다 넷플릭스에서는 사용자가 직접 평가한 평가 점수와 시청을 했는지 안 했는지, 두가지 정보만 가지고도 정확한 추천을 할 수 있다. 이에 비해 아마존에서는 위에서 말한 다양한 정보를 추천에 사용해야 하기 때문에 이런 정보를 다 수용할 수 있는 복잡한 알고리즘을 사용해야 한다.

셋째, 아마존은 넷플리스보다 취급하는 제품의 카테고리가 많다. 단순히 제품의 가지 수가 많아서 데이터가 커지는 것과는 다른 문제이다. 사용자는 카테고리가 다른 제품에 대해서는 사용 패턴도 다르고 목적도 다르다. 예를 들어 디지털 카메라와 USB 메모리는 구매 주기도 다르고 구매 결정의 기준도 다르다. 골프공과 연극표는 구매하는 목적이 다르고 구매 기준도 다르다. 이러한 차이 때문에 한 카테고리의 제품에 대한 정보는 다른 카테고리의 제품 추천에 적용하는 것은 주의해야 하고, 만일 적용한다면 카테고리 간의 관계를 고려해야 한다.
