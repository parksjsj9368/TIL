```python
# u.user 텍스트 파일을 DataFrame으로 읽기 
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('./Downloads/데이터/u.user', sep='|', names=u_cols, encoding='latin-1')
users = users.set_index('user_id')
users.head()
```

```python
# u.item 파일을 DataFrame으로 읽기
i_cols = ['movie_id', 'title', 'release date', 'video release date', 'IMDB URL', 
          'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 
          'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 
          'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies = pd.read_csv('./Downloads/데이터/u.item', sep='|', names=i_cols, encoding='latin-1')
movies = movies.set_index('movie_id')
movies.head()
```

```python
# u.data 파일을 DataFrame으로 읽기
r_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv('./Downloads/데이터/u.data', sep='\t', names=r_cols, encoding='latin-1') 
ratings = ratings.set_index('user_id')
ratings.head()
```

# 2.2 인기제품 방식

**인기 있는 제품 추천**

- 개별 사용자에 대한 정보가 없는 경우
- 정확도에 관계없이 가장 간단한 추천을 제공해야 하는 상황
<br>

- movie별 평균 ratings를 구해, 상위 영화를 출력한다

```python
movie_mean = ratings.groupby(['movie_id'])['rating'].mean()

# Best-seller 추천1
def recom_movie1(n_items):
    movie_sort = movie_mean.sort_values(ascending=False)[:n_items]
    recom_movies = movies.loc[movie_sort.index]
    recommendations = recom_movies['title']
    return recommendations

recom_movie1(5)
```

```python
# Best-seller 추천2 : 간단한 버전
def recom_movie2(n_items):
   return movies.loc[movie_mean.sort_values(ascending=False)[:n_items].index]['title']

recom_movie2(5)
```

# 2.3 추천 시스템의 정확도 측정

- y_true : 각 사용자가 평가한 모든 영화의 평점
- y_pred : 해당 사용자가 평가한 영화의 평점평균, 평점평균이 해당 영화의 예측값.
<br>

- 실제 고객이 본 영화의 rating과 사용자들이 메긴 평균 rating을 비교

```python
# 정확도 계산 (RMSE가 작을수록 정확한 추천 시스템)
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))

rmse = []
for user in set(ratings.index):
    y_true = ratings.loc[user]['rating']
    y_pred = movie_mean[ratings.loc[user]['movie_id']]
    accuracy = RMSE(y_true, y_pred)
    rmse.append(accuracy)
print(np.mean(rmse))
```

# 2.4 사용자 집단별 추천

- 사용자들을 비슷한 특성의 사람들을 묶은 소집단으로 만든 다음에 각 집단의 평점평균을 바탕으로 추천하는 것

- 비슷한 사람을 결정하는 기준 : 성별, 나이, 직업 등

```python
users = users.reset_index()
# timestamp 제거 
ratings = ratings.reset_index().drop('timestamp', axis=1)
# movie ID와 title 빼고 다른 데이터 제거
movies = movies.reset_index()[['movie_id', 'title']]
```

```python
# train, test set 분리
from sklearn.model_selection import train_test_split

x = ratings.copy()
y = ratings['user_id']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y)
```

```python
# 정확도(RMSE)를 계산하는 함수
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))

# 모델별 RMSE를 계산하는 함수 
def score(model):
    id_pairs = zip(x_test['user_id'], x_test['movie_id'])
    y_pred = np.array([model(user, movie) for (user, movie) in id_pairs])
    y_true = np.array(x_test['rating'])
    return RMSE(y_true, y_pred)
```

```python
# train 데이터로 Full matrix 구하기 
rating_matrix = x_train.pivot(index='user_id', columns='movie_id', values='rating')
rating_matrix
```

- 전체 평균으로 예측

```python
train_mean = x_train.groupby(['movie_id'])['rating'].mean() # 전체 사용자의 영화별 평균 평점

# 전체 평균으로 예측치를 계산하는 기본 모델
def best_seller(user_id, movie_id):
    try:
        rating = train_mean[movie_id]
    except:
        rating = 3.0
    return rating

train_mean = x_train.groupby(['movie_id'])['rating'].mean()
score(best_seller)
```

```python
# Full matrix를 사용자 데이터와 merge
merged_ratings = pd.merge(x_train, users)
users = users.set_index('user_id')
merged_ratings

# gender별 평점평균 계산
g_mean = merged_ratings[['movie_id', 'sex', 'rating']].groupby(['movie_id', 'sex'])['rating'].mean()
g_mean
```

- gender 기준 예측

```python
# gender별 평균을 예측치로 돌려주는 함수 
def cf_gender(user_id, movie_id):
    if movie_id in rating_matrix:
        gender = users.loc[user_id]['sex']
        if gender in g_mean[movie_id]:
            gender_rating = g_mean[movie_id][gender]
        else:
            gender_rating = 3.0
    else:
        gender_rating = 3.0
    return gender_rating

score(cf_gender)
```

- 사용자의 성별에 따른 추천이 정확도를 개선하지 못한다.

- 물론 다른 변수(나이, 직업 등)를 사용하면 경과가 달라질 수 있다.
